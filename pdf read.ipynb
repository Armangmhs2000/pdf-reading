```python
path='IELTS TERMINAL VERSION.pdf'
with open(path,"rb") as file:
    reader=PyPDF2.PdfReader(file)
    pages_to_read = [150,153,154]
    for i in pages_to_read:
        page = reader.pages[i]
        text = page.extract_text()
        print(f"--- Page {i+1} ---")
        print(text)
        print("\n")
#how to read specific pages
```

```python
import PyPDF2
path = 'IELTS TERMINAL VERSION.pdf'
with open(path, "rb") as file:
    reader = PyPDF2.PdfReader(file)
    start_page = 160   
    end_page = 165  
    for i in range(start_page, end_page):
        page = reader.pages[i]
        text = page.extract_text()
        print(f"--- Page {i+1} ---")
        print(text)
        print("\n")
#how to read pages in the range
```
